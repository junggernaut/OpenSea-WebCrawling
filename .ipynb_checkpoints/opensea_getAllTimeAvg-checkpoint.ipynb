```python
from selenium import webdriver as wd
from bs4 import BeautifulSoup
import time
import requests
import urllib.request
from selenium.webdriver.common.by import By
import csv


dr = wd.Chrome('chromedriver')
url = "https://opensea.io/rankings?sortBy=total_volume&chain=klaytn&category=collectibles"
dr.get(url)
time.sleep(3) 


# while True:
#     reslist=[]
#     while True: 
#         before_location = dr.execute_script("return window.pageYOffset") 
#         time.sleep(3)
#         elements = dr.find_elements(By.XPATH, value='//a[@class="styles__StyledLink-sc-l6elh8-0 ekTmzq Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 Itemreact__ItemBase-sc-1idymv7-0 styles__FullRowContainer-sc-12irlp3-0 bnxsDk jYqxGr dCVDRE lcvzZN fresnel-greaterThanOrEqual-xl"]')  
#         for eachelem in elements:
#             divelem = eachelem.find_elements(By.XPATH, value='*')
#             collectlist = []
#             separator = ";"
#             try:
#                 rank = divelem[0].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
#                 collectlist.append(rank)
#                 name = divelem[0].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX Ranking--collection-name-overflow"]').text
#                 collectlist.append(name)
#             except:
#                 collectlist.append("null")

#             try:
#                 img = divelem[1].find_element(By.XPATH, value='.//button[@class="UnstyledButtonreact__UnstyledButton-sc-ty1bh0-0 btgkrL"]')
#                 imgsrc = img.get_attribute("aria-label")
#                 if imgsrc == "ETH logo":
#                     collectlist.append("ETH")
#                 elif imgsrc == "KLAY logo":
#                     collectlist.append("KLAY")
#                 else:
#                     print("here")
#                 vol = divelem[1].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
#                 collectlist.append(vol)
#             except:
#                 print("null")
#                 collectlist.append("null")
#                 collectlist.append("null")

#             try:
#                 day = divelem[2].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
#                 collectlist.append(day)            
#             except:
#                 collectlist.append("null")

#             try:
#                 week = divelem[3].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
#                 collectlist.append(week)            
#             except:
#                 collectlist.append("null")

#             try:
#                 floor = divelem[4].find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
#                 collectlist.append(floor)   
#             except:
#                 collectlist.append("null")

#             try:
#                 owner = divelem[5].find_element(By.XPATH, value='.//p[@class="Blockreact__Block-sc-1xf18x6-0 Textreact__Text-sc-1w94ul3-0 styles__StatText-sc-12irlp3-3 kWYeTM heRNcW"]').text
#                 collectlist.append(owner)   
#             except:
#                 collectlist.append("null")

#             try:
#                 item = divelem[6].find_element(By.XPATH, value='.//p[@class="Blockreact__Block-sc-1xf18x6-0 Textreact__Text-sc-1w94ul3-0 styles__StatText-sc-12irlp3-3 kWYeTM heRNcW"]').text
#                 collectlist.append(item) 
#             except:
#                 collectlist.append("null")
#             result = separator.join(collectlist)
#             if not result == ";;;;;;;;":
#                 reslist.append(collectlist)
#         dr.execute_script("window.scrollTo(0,{})".format(before_location + 2500))
#         after_location = dr.execute_script("return window.pageYOffset")
#         if before_location == after_location: 
#             break
#     result = list(map(list,set(map(tuple,reslist))))
#     with open('all_nft.csv', 'a',newline='', encoding='UTF-8') as f: 
#         write = csv.writer(f) 
#         write.writerows(result) 


#     buttons = dr.find_elements(By.XPATH, value='//div[@class="Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 krGFYb jYqxGr"]/*')

#     if buttons[1].get_property('disabled'):
#         break
#     else:
#         buttons[1].click()

# while True:
#     reslist=[]
#     linklist=[]
#     while True: 
#         before_location = dr.execute_script("return window.pageYOffset") 
#         time.sleep(3)
#         elements = dr.find_elements(By.XPATH, value='//a[@class="styles__StyledLink-sc-l6elh8-0 ekTmzq Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 Itemreact__ItemBase-sc-1idymv7-0 styles__FullRowContainer-sc-12irlp3-0 bnxsDk jYqxGr dCVDRE lcvzZN fresnel-greaterThanOrEqual-xl"]') 
#         for eachelem in elements:
#             linklist.append(eachelem.get('href'))
#         dr.execute_script("window.scrollTo(0,{})".format(before_location + 2500))
#         after_location = dr.execute_script("return window.pageYOffset")
#         if before_location == after_location: 
#             break
#     reslist = list(map(list,set(map(tuple,linklist))))


#     buttons = dr.find_elements(By.XPATH, value='//div[@class="Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 krGFYb jYqxGr"]/*')

#     if buttons[1].get_property('disabled'):
#         break
#     else:
#         buttons[1].click()

reslist=[]
linklist=[]
while True:
    while True: 
        before_location = dr.execute_script("return window.pageYOffset") 
        time.sleep(2)
        elements = dr.find_elements(By.XPATH, value='//a[@class="styles__StyledLink-sc-l6elh8-0 ekTmzq Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 Itemreact__ItemBase-sc-1idymv7-0 styles__FullRowContainer-sc-12irlp3-0 bnxsDk jYqxGr dCVDRE lcvzZN fresnel-greaterThanOrEqual-xl"]') 
        for eachelem in elements:
            link = eachelem.get_attribute("href")
            rankname = eachelem.find_elements(By.XPATH, value='*')[0]
            rank = rankname.find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
            name = rankname.find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX Ranking--collection-name-overflow"]').text
            if not(rank == ''):
                linklist.append((int(rank), name, link))
        dr.execute_script("window.scrollTo(0,{})".format(before_location + 2500))
        after_location = dr.execute_script("return window.pageYOffset")
        if before_location == after_location: 
            break
    prevNextButtons = dr.find_elements(By.XPATH, value='//div[@class="Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 krGFYb jYqxGr"]/*')
    if prevNextButtons[1].get_property('disabled'):
        break
    else:
        prevNextButtons[1].click()


linklist = list(set(linklist))
linklist.sort(key = lambda x : x[0])
print(len(linklist))

for collection in linklist:
    dr.get(collection[2]+"?tab=activity")
    time.sleep(3)
    infobox = dr.find_element(By.XPATH, value='//div[@class="Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 InfoContainerreact__InfoContainer-sc-15x3z7c-0 CollectionStatsBarreact__Container-sc-8gdi9o-0 elqhCm jYqxGr fprnFG cWlCaZ"]')
    floorbox = infobox.find_elements(By.XPATH, value='*')[2]
    floor = floorbox.find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text

    try:
        button = dr.find_element(By.XPATH, value = '//div[@class="Inputreact__StyledContainer-sc-3dr67n-0 iAeYiQ Selectreact__SelectInput-sc-1shssly-0 cJLIjY"]')
    except:
        print((collection[0], collection[1], floor, "no activity"))
        reslist.append([collection[0], collection[1], floor, "no activity"])
        continue
    button.click()
    time.sleep(1)
    divelem = dr.find_elements(By.XPATH, value='.//li[@class="elements__StyledListItem-sc-197zmwo-0 QbTKh"]')
    divelem[6].click()
    time.sleep(3)
    
    try:
        alltimeavg = dr.find_element(By.XPATH, value='//div[@class="PriceHistoryStats--value"]').text
    except:
        print((collection[0], collection[1], floor, "no All Time Avg"))
        reslist.append([collection[0], collection[1], floor, "no All Time Avg"])
        continue
    print((collection[0], collection[1], floor, alltimeavg[1:]))
    reslist.append([collection[0], collection[1], floor, alltimeavg[1:]])
print(len(reslist))

with open('alltimeavg.csv', 'a', newline='', encoding='UTF-8') as f: 
    write = csv.writer(f) 
    write.writerows(reslist) 

dr.get("https://opensea.io/collection/age-of-zen-official-2?tab=activity")
time.sleep(3)
infobox = dr.find_element(By.XPATH, value='//div[@class="Blockreact__Block-sc-1xf18x6-0 Flexreact__Flex-sc-1twd32i-0 InfoContainerreact__InfoContainer-sc-15x3z7c-0 CollectionStatsBarreact__Container-sc-8gdi9o-0 elqhCm jYqxGr fprnFG cWlCaZ"]')
floorbox = infobox.find_elements(By.XPATH, value='*')[2]
floor = floorbox.find_element(By.XPATH, value='.//div[@class="Overflowreact__OverflowContainer-sc-7qr9y8-0 jPSCbX"]').text
print(floor)

button = dr.find_element(By.XPATH, value = '//div[@class="Inputreact__StyledContainer-sc-3dr67n-0 iAeYiQ Selectreact__SelectInput-sc-1shssly-0 cJLIjY"]')
button.click()
time.sleep(1)
divelem = dr.find_elements(By.XPATH, value='.//li[@class="elements__StyledListItem-sc-197zmwo-0 QbTKh"]')
divelem[6].click()

```
